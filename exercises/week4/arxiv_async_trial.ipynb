```python
import httpx

url = "https://export.arxiv.org/api/query?search_query=au:del_maestro+AND+submittedDate:[202301010600+TO+202401010600]"

async def arxiv_runner():
    client = httpx.AsyncClient()
    r = await client.get(url=url)
    # print(r.text)
    print(r.json)

await arxiv_runner()
```
