**Exercise 3: Basic Groupby - Fraud vs Normal Analysis**

**Part A: Group by Fraud Status**

  Group all transactions by Class column
    Calculate for each group:

    How many transactions? (count)
    What's the average amount? (mean)
    What's the total spending? (sum)
    How spread out are amounts? (std)

```python
# Part A - How many transaction per group
cc_class_group = cc_txn_df.groupby(by='Class')
cc_class_group.size()
```

```python
# Part A - Average amount per group
cc_class_group['Amount'].mean()
```

```python
# Sum of Amount per group
cc_class_group['Amount'].sum()
```

```python
# Spread out of Amounts : Standard Deviation - Std
cc_class_group['Amount'].std()
```

```python
# Aggregate of all
cc_class_group['Amount'].agg(['mean','sum','std'])
```

**Part B: Create Amount Categories**

  Create bins:

    Micro: $0-$10
    Small: $10-$50
    Medium: $50-$200
    Large: $200-$500
    Very Large: $500+

  Group by these bins
  Calculate total spending per bin
  Sort to find top 5

```python
bins=[0,10,50,200,500,float('inf')]
labels=['Micro','Small','Medium','Large','Very Large']
cc_txn_df['Amount_Range']=pd.cut(cc_txn_df['Amount'],bins=bins,labels=labels, include_lowest=True)
# cc_txn_df['Amount_Range'].value_counts()

```

```python
cc_txn_df
```

```python
cc_amount_group = cc_txn_df.groupby('Amount_Range')
cc_amount_group['Amount'].sum()
```

```python
cc_amount_group['Amount'].sum().sort_values(ascending=False)
```
