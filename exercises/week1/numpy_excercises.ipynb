## **Practice Challenges**

Try these to solidify your understanding:

1. **Matrix chain**: Create 3 random matrices A(3x4), B(4x5), C(5x2). Compute (A @ B) @ C and verify the shape.

2. **Normalize 2D array**: Given a 2D array, normalize it so each row has mean=0 and std=1 (use broadcasting).

3. **Pairwise distances**: Given two sets of points (shape Nx2 and Mx2), compute all pairwise distances (result NxM) without loops.

4. **Simulate dice rolls**: Simulate rolling 2 dice 10,000 times and plot histogram of sums (use random generation).

5. **Vectorize complex condition**: Given array, replace values: if x < 0 → 0, if 0 ≤ x < 10 → x², else → 100. Do it without loops.

**Hints:**
- #2: Use `axis` parameter in mean/std, and broadcasting for subtraction/division
- #3: Think about broadcasting with shapes (N, 1, 2) and (1, M, 2)
- #5: Use `np.select()` or chained `np.where()`
