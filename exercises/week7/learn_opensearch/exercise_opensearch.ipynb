```python
from opensearchpy import OpenSearch

from concepts.learn_opensearch.learn_openserach_client import response

client = OpenSearch(hosts=[{"host":"localhost", "port":"9200"}],
                    http_auth=("admin","YourStrongPassword123!"),
                    use_ssl=True,
                    verify_certs=False,
                    ssl_show_warn=False
)

client.info()
```

1. Create Index

Index name: papers

Fields:

title → text type (full-text search)

abstract → text type (full-text search)


Settings: Use defaults




```python
index_name = 'papers'
mapping={
    'settings':{
        'number_of_shards':1,
        'number_of_replicas':0
    },
    'mappings':{
        'properties':{
            'title':{'type':'text'},
            'abstract':{'type':'text'}
        }
    }
}

response = client.indices.create(index=index_name, body=mapping)
response
```

2. Index Document

Document ID: paper1

3. Search

Query: Find papers with "machine learning" in title

Query type: match query on title field

Display: Print title and relevance score for each result


```python
query={
    'query':{
        'match':{
            'title':'machine learning'
        }
    }
}

result = client.search(index=index_name, body=query)
result['hits']['hits'][0]['_score']
```
