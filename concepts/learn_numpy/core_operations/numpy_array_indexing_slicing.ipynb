**Slicing of an area changes the original array as well, even if we assign the sliced array to a new variable and then change only the sliced array.**

```python
orig_array1 = np.arange(11)
orig_array1
```

```python
slice_of_array = orig_array1[:5]
slice_of_array
```

```python
slice_of_array[:4]=90
slice_of_array
```

```python
orig_array1
```

**To safely slice (and prevent original array from changing), use copy function to craete slice using copy**

```python
orig_array2=np.arange(11)
orig_array2
```

```python
slice_by_copy = orig_array2.copy()
slice_by_copy
```

```python
slice_by_copy[:5] = 85
slice_by_copy
```

```python
orig_array2
```
