```python
import pandas as pd
```

**data frame comapred to dictornary :**
    dataframe column becomes the key of dictionary

```python
# sample dictonary
people = {
    "first":["Jane", "Nancy", "David"],
    "last":["Jones", "Drew", "Whittman"],
    "email":["jane.jones@email.com", "nancy.drew@email.com", "david.whittman@email.com"]
}
people
```

```python
# default row-lables
people_df = pd.DataFrame(people)
people_df
```

```python
people_df.columns
```

```python
# rename columns
people_df.columns=('first_name','last_name','email_id')
people_df
```

```python
people_df.columns= [col.upper() for col in people_df.columns]
people_df
```

```python
people_df.columns= people_df.columns.str.replace("_"," ")
people_df

```

```python
people_df.rename(columns={'FIRST NAME':'first'},inplace=True)
people_df
```

```python
people_df.loc[2]=['New David','New Whittman',"ndnw@abc.com"]
people_df
```

```python
people_df.loc[2,['first']]=['Old David']
people_df
```

```python
people_df.at[2,'first']='Old David'
```
