```python
import gensim
import gensim.downloader  # <--- Add this line

model = gensim.downloader.load("glove-wiki-gigaword-50")
model["tower"]
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Your vector (assuming it's in a variable called 'vector')
vector = model["tower"]

# Create a bar chart
plt.figure(figsize=(12, 4))
plt.bar(range(len(vector)), vector)
plt.title("Vector Representation of 'tower' (50 Dimensions)")
plt.xlabel("Dimension Index")
plt.ylabel("Value")
plt.show()
```

```python
from sklearn.decomposition import PCA

words = ["tower","king", "green","field","building", "sky", "river", "man", "woman","lake", "space", "girl", "blue", "child", "baby","nino", "feminino","him","her","ello", "ella"]
vectors = [model[w] for w in words]

# Reduce from 50D to 2D
pca = PCA(n_components=2)
vectors_2d = pca.fit_transform(vectors)

# Plot
plt.figure(figsize=(8, 8))
plt.scatter(vectors_2d[:, 0], vectors_2d[:, 1])

for i, word in enumerate(words):
    plt.annotate(word, xy=(vectors_2d[i, 0], vectors_2d[i, 1]))

plt.title("Word Embeddings Projected to 2D (PCA)")
plt.show()
```
