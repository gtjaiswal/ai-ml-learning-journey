```python
import instructor
from openai import OpenAI
from pydantic import BaseModel

class City(BaseModel):
    name: str
    country: str
    description: str

class Itinerary(BaseModel):
    cities: list[City]

client = instructor.from_openai(
    OpenAI(
        base_url="http://localhost:11434/v1",
        api_key="ollama"
    ),
    mode=instructor.Mode.JSON,
)

payload = [
    {"role": "user", "content": "List 5 cities with name, country, and brief description"}
]

# Instructor returns a parsed Pydantic object directly!
city_list = client.chat.completions.create(
    model="llama3.2:latest",
    stream=False,
    messages=payload,
    response_model=Itinerary,
)

# city_list is already an Itinerary object - just use it!
print(f"âœ… Instructor returned: {type(city_list)}")
print(f"\nFound {len(city_list.cities)} cities:\n")

for city in city_list.cities:
    print(f"  ğŸ“ {city.name}, {city.country}")
    print(f"     {city.description}\n")
```
