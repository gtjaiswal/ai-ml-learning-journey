```python
import httpx
url = 'https://api.github.com/events'
response= httpx.get(url)
response.json()
```

```python
with httpx.Client() as client:
    response =client.get(url)
    print(response.json())
```

```python
async with httpx.AsyncClient() as client:
    response = await client.get(url)
    print(response.json())
```
