```python
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsRegressor
import matplotlib.pylab as plt
```

```python
X, y = load_iris(return_X_y=True)
```

```python
mod = KNeighborsRegressor()
mod.fit(X, y)
pred = mod.predict(X)
pred.size
```

```python
plt.scatter(pred, y)
```
