```python
import pandas as pd
```

**data frame comapred to dictornary :**
    dataframe column becomes the key of dictionary

```python
# sample dictonary
people = {
    "first":["Jane", "Nancy", "David"],
    "last":["Jones", "Drew", "Whittman"],
    "email":["jane.jones@email.com", "nancy.drew@email.com", "david.whittman@email.com"]
}
people
```

# Create Data Frame from dictonary
**key becomes column (and coumn-lables)**
**row-lables are given as index : is by default given as 0,1,2......**

```python
# default row-lables
people_df = pd.DataFrame(people)
people_df
```

```python
# custom row-labes
people_df_rowlables = pd.DataFrame(people)
people_df_rowlables=people_df_rowlables.rename(index={0: 'customer 1', 1: 'customer 2', 2: 'customer 3'})
people_df_rowlables
```

```python
people_df_rowlables_1 = pd.DataFrame({'Bob': ['I liked it.', 'It was awful.'], 
              'Sue': ['Pretty good.', 'Bland.']},
             index=['Product A', 'Product B'])
people_df_rowlables_1
```

```python
#  "Series" - series is only with one column. If you give more than one column it becomes a dataFrame
type(people_df['email'])
```

```python
people_series = pd.Series([30,40,50])
people_series
```

```python
type(people_series)
```

```python
# Series have row-label, but only one column so no column lable, just the name of the series
people_series = pd.Series([30,40,50],index=['Type A','Type B','Type C'])
people_series
```

```python
people_series = pd.Series([30,40,50],index=['Type A','Type B','Type C'],name="Product")
people_series
```

**The Series and the DataFrame are intimately related. It's helpful to think of a DataFrame as actually being just a bunch of Series "glued together"**
