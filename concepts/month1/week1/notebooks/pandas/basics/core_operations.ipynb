```python
# By default Pandas craete a default index calling when a file is read. To mak eof the existing columns from the files as index, specify col_index
data_frame = pd.read_csv("../../data/stack-overflow-2025/survey_results_public.csv",index_col=0)
data_frame
```

```python
# iloc - location accessor by index : get only rows with index 2 and 6 and column with index 10 - always integers
data_frame.iloc[[2,6],[10]]
```

```python
# loc - location accessor by lable : get the row with row-lable. this is not the dafault index, but the label of the row/column
data_frame.loc[[3],["Age"]]
```
