```python
country_respondents=df['Country'].value_counts()
country_respondents
```

```python
python_df = pd.concat([country_respondents,country_uses_python],axis='columns', sort=False)
python_df
```

```python
python_df.rename(columns={'count':'Num of Respondents','LanguageHaveWorkedWith':'Num who knows Python'}, inplace=True)
python_df
```

```python
python_df['%KnwPython']=python_df['Num who knows Python']/python_df['Num of Respondents']*100
python_df.sort_values(by='%KnwPython', ascending=False)
```
