## Indexing in Pandas:
**- .iloc :** 
 - numeric position based indexing (need to give the row or column positional index)
 - first element of the range is included and last excluded - [0:1000] will return 1000 elements (Std Python behvaior)
**- .loc :** 
 - label based indexing (need to give the row or column name)
 - both first and last elemets of the range are included - [0:1000] will return 1001 elements ( behavior diff from that of standard python), so use [0:999] for 1000 elements

## ways of selecting:
examples :
first_description = reviews.loc[0,'description']
first_description = reviews.description[0]
first_description = reviews.description.iloc[0]
first_description = reviews.description.loc[0]

## Manipulating the Index
**setting index**
df.set_index(<new index>) - will not change the underlying data frame
df.set_index(<new index>,inplace = true)

**re-setting index**
df.reset_index(inplace = true)

**sorting sf by index**
df.sort_index()
df.sort_index(inplace=true)

## Conditional Selection
**single condition**
df.<column_name/row_name/index> == <value to be matched> - give true/false
df.loc[condition above....]

**multiple coditions using logical operators**
df.loc [condition 1 & condition 2]

**conditions using pandas built-in functions**
- **isin** df.<column_name/row_name/index>.isin(<list to be matched>) - list...
- **notnull**
-**negate filter** ~ sign
