## 
X - Data needed to make prediction, eg : properties of house or info about house
y - what is to be predicted, eg : house price

```python
# load the data
from sklearn.datasets import load_iris
X, y = load_iris(return_X_y=True)
X
```

```python
y
```

```python
# select and create the model
from sklearn.neighbors import KNeighborsRegressor
knn_mod = KNeighborsRegressor()
knn_mod
```

```python
# not fitted so cant predict - below will error out
# knn_mod.predict(X)

knn_mod.fit(X,y)
```

```python
# now predict as fitted
knn_prediction = knn_mod.predict(X)
knn_prediction.size
```

```python
# select and create another model
from sklearn.linear_model import LinearRegression
lr_mod = LinearRegression()
lr_mod
```

```python
lr_mod.fit(X,y)
```

```python
lr_prediction = lr_mod.predict(X)
lr_prediction.size
```
