```python
import xml.etree.ElementTree as ET

tree=ET.parse("library.xml")
```

```python
# get root
root = tree.getroot()
print("root =",root)
print("root.tag =",root.tag)
print("root.keys() =",root.keys())
print("root.items()=",root.items())
print("root.attrib =",root.attrib)
print("root.text =",root.text)
```

1. **.tag →** the element’s name (library) ✅ correct.
2. **.attrib →** a dictionary of attributes of the element. Your <library> has no attributes, so {}.
3. **.keys() →** list of attribute names. Empty {} → empty list [].
4. **.items() →** list of (key, value) pairs of attributes. Empty {} → empty list [].
5. **.text →** the text directly inside the element, before any child elements.

* The library element has no text outside the <book> child, so .text is blank.
* The <book> elements are children, not attributes or text of <library>.

```python
# get child elements
print("root.get(\"book\")",root.get("book"))
print("root.find(\"book\")=",root.find("book"))
print("root.findall(\"book\")=",root.findall("book"))
```

```python
for book in root.findall("book"):
    # print("book.text=",book.text)
    # print("book.attrib=",book.attrib)
    # print("book.keys=",book.keys())
    # print("book.items()=",book.items())
    # print("book.tag=",book.tag)
    # print("book.tail=",book.tail)
    print(book.find("title").text)
    print(book.get("id"))
    print(book.find("author").text)
```

```python
tree= ET.parse("nested_books.xml")
root = tree.getroot()
books = root.findall("book")
for book in books:
    print("book= ",book.find("title").text)
    details = book.find("details")
    # print(details)
    print("\tpublisher= ", details.find("publisher").text)
    print("\tyear= ",details.find("year").text)
```

```python
tree= ET.parse("books_attributes.xml")
root = tree.getroot()
for book in root.findall("book"):
    if (int(book.get("published")))>2020: print(f"title= {book.find("title").text}, publisher= {book.find("publisher").text}")

```
